**1. What is Exception in Python?**

- In Python programming, an exception is an event that occurs during the execution of a program and disrupts the normal flow of instructions.

- It is an indication that something unexpected or erroneous has happened, which prevents the program from continuing its normal execution.

**Exceptions can occur due to various reasons:**

- **Programming errors:**

  These are mistakes made by the programmer, such as dividing a number by zero or accessing an invalid array index.

- **External factors:**

  These include issues outside the control of the program, such as hardware failures, network errors, or file system problems.

- **Invalid user input:**

  When the user provides input that doesn't confirm to the expected format or violates certain constraints, an exception may be raised.

- **Resource limitations:**

  If the program exceeds the available system resources, such as memory or file handles, an exception can be triggered.

**What is Exception Handling in Python ?**

- Exception handling in Python refers to the process of handling and managing errors or exceptions that may occur during the execution of a program.
- It allows you to anticipate and gracefully deal with errors, ensuring that your program can recover from unexpected situations rather than abruptly terminating.

**2. Handling Exception with syntax in Python and example**

**To handle exceptions in programming, we can follow these steps:**

- Wrap the code that might raise an exception within a try block.

- Specify the type of exception you want to handle using an except block.

- Write the code that should be executed if the specified exception occurs.

- Optionally, you can have multiple except blocks to handle different types of exceptions.

- Finally, you can include an optional else block that will be executed if no exception occurs.

- Optionally, you can include a finally block that will be executed regardless of whether an exception occurred or not. This block is used for cleanup operations.

**Syntax:**


```python
try:
    # Code that may raise an exception
    # ...
except ExceptionType1:
    # Handle ExceptionType1
    # ...
except ExceptionType2:
    # Handle ExceptionType2
    # ...
else:
    # Code to be executed if no exception occurs
    # ...
finally:
    # Code to be executed regardless of whether an exception occurred or not
    # ...
```

- **Try:** This block will test the excepted error to occur
- **Except:**  Here you can handle the error
- **Else:** If there is no exception then this block will be executed
- **Finally:** Finally block always gets executed either exception is generated or not

**Handling exceptions is important for several reasons:**

- **Error reporting and user experience**:

  Handling exceptions allows us to provide meaningful error messages or instructions to users when something goes wrong. Instead of crashing or showing a cryptic error message, you can gracefully handle exceptions and guide users on how to resolve or mitigate the issue.

- **Program stability and robustness:**

  Exception handling helps prevent our program from abruptly terminating when unexpected errors occur. By handling exceptions, you can catch and recover from errors, allowing the program to continue executing and providing a more stable experience.

- **Debugging and troubleshooting:**

  When an exception occurs, the program can provide information about the cause of the error, such as the type of exception and the line of code where it occurred. This information is valuable for debugging and troubleshooting, as it helps identify and fix issues in the code.

- **Resource management:**

  Exception handling is essential for properly releasing resources like file handles, network connections, or database connections. By handling exceptions and including cleanup operations in the finally block, we can ensure that resources are properly managed and released, even if an exception occurs.


**Example:**

Program to display error message when a number is divided by zero

```python
try:
    dividend = int(input("Enter the dividend: "))
    divisor = int(input("Enter the divisor: "))
    result = dividend / divisor
    print("Result:", result)
except ValueError:
    print("Invalid input. Please enter a valid number.")
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
except Exception as e:
    print("An error occurred:", str(e))
else:
    print("Division performed successfully.")
finally:
    print("Program execution complete.")
```

**In this example:**

- The try block contains the code that may raise exceptions. We attempt to divide the dividend by the divisor.

- The except ValueError block catches a ValueError if the user enters an invalid input (e.g., a non-numeric value).

- The except ZeroDivisionError block catches a ZeroDivisionError if the user attempts to divide by zero.

- The except Exception block is a catch-all for any other exceptions that may occur. It prints a general error message along with the exception details.

- The else block is executed if no exception occurs. It indicates a successful division.

- The finally block is always executed, regardless of whether an exception occurred or not. It provides a message to indicate the completion of program execution.

```python
try:
    # Code that may raise exceptions
except:
    # Exception handling code
```

- It will catch exceptions of any type, including built-in exceptions like ValueError, TypeError, or ZeroDivisionError, as well as custom exceptions.

**Example of using an except clause with no exceptions specified**

```python
try:
    result = 10 / 0
except:
    print("An exception occurred")
```

**In this example:**

The try block attempts to divide 10 by zero, which raises a ZeroDivisionError. The catch-all except clause catches the exception and prints a generic error message.

**The except Clause with Multiple Exceptions**

- In Python, we can use an except clause with multiple exceptions to handle different types of exceptions in a single block of code.
- This allows you to provide specific handling for each exception type and customize your error handling based on the type of exception raised.

**Here's the syntax for using an except clause with multiple exceptions:**

```python
try:
    # Code that may raise exceptions
except ExceptionType1:
    # Exception handling for ExceptionType1
except ExceptionType2:
    # Exception handling for ExceptionType2
except ExceptionType3:
    # Exception handling for ExceptionType3
...
except:
    # Default exception handling for other exceptions
```

**Here's an example of using an except clause with multiple exceptions:**

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid value")
except Exception as e:
    print("An exception occurred:", str(e))
```

**In this example:**

- The try block attempts to divide 10 by zero, which raises a ZeroDivisionError.
- The first except clause specifically handles ZeroDivisionError and prints a corresponding error message.
- If the exception raised were a ValueError, the second except clause would handle it.
- If the exception doesn't match any of the specified types, the last except clause (without an exception type) would catch the exception and print a generic error message.

**The try-finally Clause**

- In Python, the try-finally clause is used to define a block of code that will be executed regardless of whether an exception occurs or not.
- It ensures that certain code statements are always executed, whether an exception is raised or not, and regardless of whether an exception is caught and handled.

**The syntax of the try-finally clause is as follows:**

```python
try:
    # Code that may raise exceptions
finally:
    # Code that will always be executed
```

**Here's an example that demonstrates the use of the try-finally clause:**

```python
def divide(x, y):
    try:
        result = x / y
        print("Result:", result)
    finally:
        print("Operation completed.")

divide(10, 2)
divide(10, 0)
```

**In this example:**

- The divide function attempts to perform division between two numbers. The first call divide(10, 2) will execute successfully, resulting in the division operation being performed and the result being printed.
- The finally block will also execute, printing "Operation completed."

- The second call divide(10, 0) will raise a ZeroDivisionError since division by zero is not allowed. However, even in the presence of an exception, the finally block will still execute, ensuring that "Operation completed." is printed.
- This demonstrates that the code in the finally block is executed regardless of whether an exception occurs or not.

**exception LookupError**

- In Python, the LookupError exception class is the base class for exceptions that occur when a lookup or indexing operation fails.
- It is a subclass of the Exception class and serves as the superclass for more specific lookup-related exception classes.

**Some commonly used exception classes derived from LookupError include:**

- **IndexError:**
Raised when an index is out of range, typically occurring when accessing elements in a sequence like a list or tuple.

- **KeyError:** Raised when a dictionary key is not found.

- **AttributeError:** Raised when an attribute reference or assignment fails, indicating that the specified attribute does not exist.

**Here's an example that demonstrates the use of LookupError and its derived exception classes:**

```python
def process_list(lst, index):
    try:
        value = lst[index]
        print("Value at index", index, ":", value)
    except LookupError as e:
        print("LookupError occurred:", type(e).__name__)
        print("Error message:", str(e))

def process_dict(dictionary, key):
    try:
        value = dictionary[key]
        print("Value for key", key, ":", value)
    except LookupError as e:
        print("LookupError occurred:", type(e).__name__)
        print("Error message:", str(e))

def process_attribute(obj, attr):
    try:
        value = getattr(obj, attr)
        print("Value of attribute", attr, ":", value)
    except LookupError as e:
        print("LookupError occurred:", type(e).__name__)
        print("Error message:", str(e))

# LookupError: IndexError
process_list([1, 2, 3], 5)

# LookupError: KeyError
process_dict({"name": "John", "age": 25}, "city")

# LookupError: AttributeError
process_attribute("Hello", "length")
```

**In this example:**

- We have three functions that demonstrate different lookup operations.
- The process_list function attempts to access an element from a list using an index.
- The process_dict function tries to retrieve a value from a dictionary using a key.
- The process_attribute function attempts to access an attribute of an object using the getattr function.

**Customizing Exception Classes**

- When defining custom exception classes in Python, you have the flexibility to customize them by adding attributes, methods, and behavior to suit your specific needs.
- This allows you to provide more detailed information about the exceptional condition and enhance the functionality of your custom exceptions.

**Here's an example that demonstrates how to customize a custom exception class:**

```python
class CustomException(Exception):
    def __init__(self, message, code):
        super().__init__(message)
        self.code = code

    def display_error(self):
        print("CustomException occurred with code:", self.code)
        print("Error message:", self.args[0])

# Usage
def process_data(data):
    if not data:
        error_code = 1001
        error_message = "Invalid data. Empty input."
        raise CustomException(error_message, error_code)

try:
    process_data(None)
except CustomException as e:
    e.display_error()
```

**In this example**

- The CustomException class is customized with an additional __init__ method and a display_error method.

- The __init__ method is overridden to accept two parameters, message and code, in addition to the standard Exception parameters.

- It calls the __init__ method of the base class (Exception) to initialize the exception with the provided message.

- It also assigns the code parameter to the self.code attribute, allowing you to store additional information about the exception.

- The display_error method is added to the CustomException class, which displays the error information, including the code and the error message stored in the exception.

- It can be called on an instance of CustomException to provide more detailed error reporting.

- In the process_data function, we raise CustomException with an error code and message when the data parameter is empty.

- When the exception is caught, the display_error method is called on the exception object to print the error details, including the code and the error message.

**User-Defined class with Multiple Inheritance**

- In Python, we can create a user-defined class that inherits from multiple parent classes using multiple inheritance.
- Multiple inheritance allows a class to inherit attributes and methods from multiple base classes, combining their functionalities into a single derived class.

**Here's an example of a user-defined class with multiple inheritance:**

```python
class BaseClass1:
    def method1(self):
        print("BaseClass1 method1")

    def common_method(self):
        print("BaseClass1 common_method")


class BaseClass2:
    def method2(self):
        print("BaseClass2 method2")

    def common_method(self):
        print("BaseClass2 common_method")


class DerivedClass(BaseClass1, BaseClass2):
    def derived_method(self):
        print("DerivedClass derived_method")


# Usage
obj = DerivedClass()
obj.method1()  # Call method from BaseClass1
obj.method2()  # Call method from BaseClass2
obj.derived_method()  # Call method from DerivedClass
obj.common_method()  # Call method with the same name in both BaseClass1 and BaseClass2
```

**In this example:**

- We define two base classes, BaseClass1 and BaseClass2, each with their own set of methods.

**Deriving Error from Super Class Exception**

- To derive a custom error class from a superclass exception in Python, you can create a new class that inherits from the desired exception class.
- By doing this, you can define your own custom error type that inherits the behavior and attributes of the superclass exception.

**Here's an example of deriving a custom error class from a superclass exception:**

```python
class CustomError(Exception):
    def __init__(self, message):
        super().__init__(message)
class CustomValueError(CustomError, ValueError):
    pass
# Usage
def validate_input(value):
    if value < 0:
        raise CustomValueError("Input value must be positive")
try:
    validate_input(-5)
except CustomValueError as e:
    print("CustomValueError occurred:", str(e))
```

- We then create a subclass called CustomValueError that inherits from both CustomError and ValueError. The CustomValueError class doesn't add any new methods or attributes, so we use the pass statement to indicate an empty body.

- In the validate_input function, we raise the CustomValueError if the input value is negative.

**8.Advantages of Exception Handling**

- **Error handling:**

   - Exception handling provides a structured way to handle and manage errors that occur during program execution.
   - Instead of causing the program to terminate abruptly, exceptions allow you to gracefully handle errors and take appropriate actions, such as displaying error messages, logging, or performing fallback operations.

- **Robustness:**

  -  By incorporating exception handling, you can make your programs more robust and resilient. Exceptions help you anticipate and handle exceptional conditions, such as invalid input, file I/O errors, or network failures, ensuring that your program can recover from such situations and continue executing without unexpected crashes.

- **Readability and maintainability:**

  - Exception handling improves the readability and maintainability of your code. By encapsulating error handling logic within exception blocks, you separate it from the regular code flow, making the main logic more focused and easier to comprehend.

  - This separation enhances code modularity, allows for easier maintenance, and reduces the likelihood of duplicated error handling code.

- **Graceful program termination:**
  - Exceptions enable you to gracefully terminate your program when encountering unrecoverable errors.
  - By catching specific exceptions and performing appropriate cleanup operations (e.g., closing files, releasing resources), you can ensure that the program exits cleanly and avoids leaving system resources in an inconsistent or unreleased state.

- **Debugging and troubleshooting:**
  - Exception handling aids in the debugging and troubleshooting process. Exceptions provide valuable information about the cause and context of an error, including stack traces and error messages, which help pinpoint the source of the problem.
  - By handling exceptions effectively and logging relevant details, you can improve the diagnosis and resolution of issues in your code.

- **Fault tolerance and recovery:**
  - Exception handling facilitates fault tolerance and recovery strategies. When dealing with external systems, databases, or network connections, exceptions allow you to implement mechanisms for retries, fallbacks, or alternative actions in case of failures
  - This improves the overall resilience of your application and enhances its ability to handle unexpected scenarios.

**9.Disadvantages of Exception Handling**

- **Overuse and abuse:**

  - One disadvantage of exception handling is the possibility of overusing or abusing it. If exceptions are used excessively for flow control or to handle routine and expected conditions, it can make the code harder to read and maintain.
  - Exception handling should be reserved for exceptional cases and errors, while regular control flow should be handled using traditional control structures.

- **Difficulty in debugging:**
  - Exception handling can sometimes complicate the debugging process. When exceptions occur, the program flow is altered, and it can be challenging to trace the exact sequence of events that led to the exception.
  - Additionally, if exceptions are not properly handled or if the exception handling code is incorrect, it may obscure the root cause of the issue, making it harder to identify and fix the underlying problem.

- **Performance overhead:**
  - Exception handling can introduce performance overhead, especially in scenarios where exceptions are thrown frequently or in tight loops. The process of throwing and catching exceptions requires additional runtime checks and can impact the execution speed of the program.
  - In performance-critical applications or sections of code, alternative error handling mechanisms may be more suitable to minimize the impact on performance.

- **Inconsistent exception handling:**
  - In larger codebases or projects involving multiple developers, maintaining consistency in exception handling can be a challenge.
  - Different developers may handle exceptions differently, leading to inconsistency in error handling approaches.
  - Inconsistent exception handling can make it harder to understand and maintain the codebase as developers may expect different behaviors from exception handling.

- **Increased code complexity:**
  - Exception handling can introduce complexity to the codebase.
  - Handling exceptions often involves nesting try-except blocks, which can make the code more convoluted and harder to read.
  - Additionally, handling exceptions across different modules or layers of the application may require careful coordination and planning to ensure consistent error handling practices.
